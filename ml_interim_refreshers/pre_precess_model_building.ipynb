```python
df = pd.read_csv("melb_data.csv")
```

```python
df.dtypes
```

```python
a = df.select_dtypes(["float64","int64"])
```

```python
a.dtypes
```

```python
a.corr()
```

```python
a[a.BuildingArea.isna()]
```

```python
a.BuildingArea.mean()
```

```python
a.YearBuilt = a.YearBuilt.fillna(2000)
a.BuildingArea = a.BuildingArea.fillna(a.BuildingArea.mean())
```

```python
independent = a.columns[a.corr()['Price'] > 0.3]
independent
```

```python
X = a[independent]
X = X.drop('Price',axis = 1)
Y = a['Price']
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
dtree = DecisionTreeRegressor(random_state=0)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
```

```python
dtree.fit(X_train,y_train)
```

```python
pred = dtree.predict(X_test)
pred
```

```python
from math import sqrt
sqrt(mean_squared_error(y_test,pred))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
boosting_model = GradientBoostingRegressor(learning_rate=0.05,random_state=0)
boosting_model.fit(X_train,y_train)
pred = boosting_model.predict(X_test)
```

```python
from math import sqrt
sqrt(mean_squared_error(y_test,pred))
```

```python
from sklearn.ensemble import AdaBoostRegressor
ada_boost = AdaBoostRegressor(random_state=0,learning_rate=0.05)
ada_boost.fit(X_train,y_train)
pred = ada_boost.predict(X_test)
sqrt(mean_squared_error(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(n_estimators=500,random_state=0)
rf_model.fit(X_train,y_train)
pred = rf_model.predict(X_test)
sqrt(mean_squared_error(y_test,pred))
```
