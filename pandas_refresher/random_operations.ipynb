```python
numeric_types = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_data = df.select_dtypes(exclude='object')
numeric_data.head()
```

```python
columns_with_missing = numeric_data.columns[numeric_data.isna().any()]
columns_with_missing
```

```python
numeric_data1=numeric_data
```

```python
#Handling missing values
numeric_data= numeric_data.apply(lambda x: x.fillna(x.mean()),axis = 0)
```

```python
from sklearn.decomposition import PCA
pca_data = PCA(n_components=20)
```

```python
pca_transformed_data = pca_data.fit_transform(numeric_data)
```

```python
pca_df = pd.DataFrame(pca_transformed_data)
pca_df.head()
```

```python
#Feature Selection using Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
X = numeric_data.drop(['SalePrice'],axis=1)
y = numeric_data['SalePrice']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
sel = SelectFromModel(RandomForestClassifier(n_estimators = 100))
sel.fit(X_train, y_train)
```
